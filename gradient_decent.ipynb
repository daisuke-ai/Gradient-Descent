<h2>Gradient descent</h2>


```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0000001


x_list = [new_x]
slope_list = [df(new_x)]



for n in range(500):
  previous_x = new_x
  gradient = df(previous_x)
  new_x = previous_x - step_multiplier * gradient

  x_list.append(new_x)
  slope_list.append(df(new_x))

  step_size = abs(new_x - previous_x)
  if step_size < precision:
    print(n)
    break

print(f"Local minimum occurs at: {new_x}")
print(f"slope or df(x) value at this point is: {df(new_x)}")
print(f"f(x)value or cost at this point is: {f(new_x)}")
```

<h2> Gradient Descent as a Python Function </h2>

```python
import numpy as np
import matplotlib.pyplot as plt

# Previous cell code
x_2 = np.linspace(-2, 2, 1000)

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x  # Added the missing subtraction of 8*x

def gradient_descent(derivative_func, initial_guess, multiplier, precision):
    new_x = initial_guess
    x_list = [new_x]

    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        x_list.append(new_x)
        step_size = abs(new_x - previous_x)

        if step_size < precision:
            break

    return x_list, len(x_list) - 1  # Return the x_list and the number of steps taken

# Call gradient_descent function to get the list of points visited during optimization
list_x, _ = gradient_descent(dg, 0.5, 0.02, 0.0001)

# Plot function and derivative side by side
plt.figure(figsize=(20, 5))

# First chart: Cost function
plt.subplot(1, 2, 1)
plt.xlabel("X", fontsize=16)
plt.ylabel("g(x)", fontsize=16)
plt.title("Cost-Function", fontsize=16)
plt.xlim([-2, 2])
plt.ylim(0.5, 5.5)
plt.grid()
plt.plot(x_2, g(x_2), linewidth=4)
plt.scatter(list_x, g(np.array(list_x)), color="red", s=100, alpha=0.6)

# Second chart: Derivative
plt.subplot(1, 2, 2)
plt.title("Derivative")
plt.xlabel("X")
plt.ylabel("dg(x)")
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.grid()
plt.plot(x_2, dg(x_2), color="skyblue", linewidth=3)
plt.scatter(list_x, dg(np.array(list_x)), color="red", s=100, alpha=0.8)

plt.show()
```

<h2> Batch Gradient Descent with SymPy </h2>


```python
import math

def fpx(x,y):
    r = 3**(-x**2-y**2)
    return (2*x*math.log(3)*r) / (r+1)**2

def fpy(x,y):
    r = 3**(-x**2-y**2)
    return (2*y*math.log(3)*r) / (r+1)**2
```

<h2>Example 5 - Working with Data and a Real Cost Function
Mean Squared Error </h2>


$$RSS = \sum_{i=1}^{n} \big(y^{(i)} - h_\theta x^{(i)} \big)^2$$

<br>

$$MSE =\frac{1}{n} \sum_{i=1}^{n} \big(y^{(i)} - h_\theta x^{(i)} \big)^2$$



```python
# Making sample Data
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7,1)
```

```python
#Quick linear regression

regr = LinearRegression()
regr.fit(x_5, y_5)
print('Theta 0:', regr.intercept_[0])
print('Theta 1:', regr.coef_[0][0])
```

```python
plt.scatter(x_5, y_5, s=50)
plt.plot(x_5, regr.predict(x_5), color='orange', linewidth=3)
plt.xlabel("x values")
plt.ylabel("Y values")
```

```python
#lienar regression = theta0 + thera1*x

y_hat = 0.8475351486029545 + 1.2227264637835913 * x_5
print("Est values y_hat are: \n", y_hat)
print("Actual y_hat values are: \n", y_5)
```

```python
def mse(y, y_hat):
  error= sum((y_hat - y)**2)
  error = error/len(y)
  return error


error = mse(y_5, y_hat)
print(error)
```

```python
def mse(y, y_hat):
    n = len(y)  # Number of samples
    squared_errors = [(y_hat[i] - y[i])**2 for i in range(n)]
    mse = sum(squared_errors) / n
    return mse


error = mse(y_5, y_hat)
print('Manually calculated mse: ',error)
print('Built-in function from sklearn: ', mean_squared_error(y_5, y_hat))
```

<h1>Cost function MSE - Partial derivative: <h1>



<h2>$$\frac{\partial MSE}{\partial \theta_0} = -\frac{2}{n} \sum_{i=1}^{n} (y_i - (\theta_0 + \theta_1 x_i))$$<br>
$$\frac{\partial MSE}{\partial \theta_1} = -\frac{2}{n} \sum_{i=1}^{n} x_i (y_i - (\theta_0 + \theta_1 x_i))$$ </h2>



<h2> MSE and Gradient Descent </h2>

```python
def grad(x, y, thetas):
  n = y.size
  theta0_slope= (-2/n) * (sum(y - thetas[0] - thetas[1] * x))
  theta1_slope = (-2/n) * (sum(x * (y - thetas[0] - thetas[1] * x)))
  return np.array((theta0_slope[0], theta1_slope[0]))
  # return np.append(arr= theta0_slope, values=theta1_slope)
  # return np.concatenate((theta0_slope, theta1_slope), axis=0)
```
